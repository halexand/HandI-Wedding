```python
import scipy.stats
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
%matplotlib inline
```

```python
#read in data
df=pd.read_table('guestSim.txt')
#calculated bernoulli 
simNum=50000
simdata=df.LE.apply(scipy.stats.bernoulli.rvs, size=simNum)
simdata=simdata*df.Num
simdata_sum=simdata.sum(axis=0)
```

```python
fig,ax = plt.subplots(1)

pd.DataFrame(simdata_sum).hist(bins=20, ax = ax)
ax.set_title('Wedding Attendees \nProbability Distribution ({:,} simulations)'.format(simNum), fontweight='bold', fontsize = 'x-large')
ax.set_xlabel('Number of Attendees')
ax.set_yticks(np.linspace(0,8000,5))
fig.set_size_inches(6,5)
fig.savefig('WeddingSim.png', bbox='tight')
```
